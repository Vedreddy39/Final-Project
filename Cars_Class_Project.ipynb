```python
import pandas as pd

# Load dataset into a pandas dataframe
data = pd.read_csv('cars_class.csv')
```

```python
#deleting columns as per the dataset requirements
data.pop(data.columns[0])
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
# View the first 5 rows of the dataset
data.head()

# Check the dimensions of the dataset
data.shape

# Check the data types of the columns
data.dtypes

# Check for missing values
data.isna().sum()

#Based on the above exploration, we can see that there are no missing values in the dataset. Now, let's proceed with data preprocessing.
```

```python
# Drop any rows with missing values
train_data.dropna(inplace=True)
test_data.dropna(inplace=True)

# Split the data into input features and target variable
X_train = train_data.iloc[:, :-1]
y_train = train_data.iloc[:, -1]
X_test = test_data.iloc[:, :-1]
y_test = test_data.iloc[:, -1]

# Standardize the data using StandardScaler
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
"""We will try the following classifiers:"""

"""Logistic Regression
Decision Tree Classifier
Random Forest Classifier
Support Vector Classifier (SVC)
K-Nearest Neighbors (KNN) Classifier"""
```

```python
# Import required classifiers and metrics
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

# Define the classifiers
lr = LogisticRegression(random_state=42)
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)
svc = SVC(random_state=42)
knn = KNeighborsClassifier()

# Fit the classifiers on the training set
lr.fit(X_train, y_train)
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
svc.fit(X_train, y_train)
knn.fit(X_train, y_train)

# Predict the target variable on the test set
y_pred_lr = lr.predict(X_test)
y_pred_dt = dt.predict(X_test)
y_pred_rf = rf.predict(X_test)
y_pred_svc = svc.predict(X_test)
y_pred_knn = knn.predict(X_test)

# Evaluate the performance of the classifiers
print("Accuracy Scores:")
print("Logistic Regression: ", accuracy_score(y_test, y_pred_lr))
print("Decision Tree: ", accuracy_score(y_test, y_pred_dt))
print("Random Forest: ", accuracy_score(y_test, y_pred_rf))
print("SVC: ", accuracy_score(y_test, y_pred_svc))
print("KNN: ", accuracy_score(y_test, y_pred_knn))

print("\nF1 Scores:")
print("Logistic Regression: ", f1_score(y_test, y_pred_lr, average='macro'))
print("Decision Tree:", f1_score(y_test, y_pred_dt, average='macro'))
print("Random Forest: ", f1_score(y_test, y_pred_rf, average='macro'))
print("SVC: ", f1_score(y_test, y_pred_svc, average='macro'))
print("KNN: ", f1_score(y_test, y_pred_knn, average='macro'))

print("\nConfusion Matrices:")
print("Logistic Regression: \n", confusion_matrix(y_test, y_pred_lr))
print("Decision Tree: \n", confusion_matrix(y_test, y_pred_dt))
print("Random Forest: \n", confusion_matrix(y_test, y_pred_rf))
print("SVC: \n", confusion_matrix(y_test, y_pred_svc))
print("KNN: \n", confusion_matrix(y_test, y_pred_knn))
```

```python

#Based on the evaluation metrics, we can see that Logistic Regression has the highest accuracy and F1-score. 
#Therefore, we will choose Random Forest Classifier as our final model and tune its hyperparameters.
```

```python
# Set the final model as Logistic Regression
final_model = lr

# Fit the final model on the training set
final_model.fit(X_train, y_train)

# Predict the target variable on the test set
y_pred_final = final_model.predict(X_test)

# Evaluate the performance of the final model
print("Accuracy: ", accuracy_score(y_test, y_pred_final))
print("F1 Score: ", f1_score(y_test, y_pred_final, average='macro'))
print("\nConfusion Matrix: \n", confusion_matrix(y_test, y_pred_final))
```
